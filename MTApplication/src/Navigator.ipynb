```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.common.action_chains import ActionChains
from enum import Enum
```

```python
class Language(Enum):
  AMIS = '阿美族',
  PAIWAN = '排灣族'
```

```python
# Service to navigate webpages using Selenium Webdriver
class Navigator:
  def __init__(self):
    options = webdriver.EdgeOptions()
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-setuid-sandbox")
    options.add_argument("--headless")
    self.driver = webdriver.Edge(options=options)
    self.driver.maximize_window()

  # Navigates to a webpage
  def navigate(self, url: str):
    self.driver.get(url)

  # Clicks a button once if located and clickable
  # Takes a "By" type (id, class, xpath...) and identifier
  def click_button(self, by: By, id: str):
    wait = WebDriverWait(self.driver, 10)
    button = wait.until(EC.presence_of_element_located((by, id)))
    button = wait.until(EC.element_to_be_clickable((by, id)))
    if button:
      try:
        ActionChains(self.driver).move_to_element(button).click(button).perform()
      except:
        print("Button unclickable.")
    else:
      print("Button not found")

  # Exhausts a button click until there are no more pages
  def click_button_repeatedly(self, by: By, id: str):
    while self.element_exists(by, id):
      try:
        self.click_button(by, id)
      except:
        break

  def element_exists(self, by: By, id: str):
    return self.driver.find_elements(by, id)

  def get_source_html(self):
    return self.driver.page_source
```
