```python
import pandas as pd
import re

# Define the text cleaning function
def remove_punct(s):
    s = re.sub(r"\[|\]", "", s)
    s = re.sub(r"⌃", " ", s)
    s = re.sub(r"—", " ", s)
    s = re.sub(r"[\.\,\?\!\<\>;\:\’\‘]", "", s)
    s = re.sub(r'["”“]', "", s)
    s = re.sub(r'[‘’`]', "'", s)
    s = re.sub(r"-", " ", s)
    s = s.lower()
    return s

# Function to load, clean, and save the CSV
def clean_csv(input_file_path, output_file_path):
    # Load the CSV file
    df = pd.read_csv(input_file_path)

    # Apply the cleaning function to each element in the dataframe
    for column in df.columns:
        df[column] = df[column].astype(str).apply(remove_punct)

    # Save the cleaned data to a new CSV file
    df.to_csv(output_file_path, index=False)
    print("File has been cleaned and saved.")

# Example usage (update paths as necessary)
input_file_path = '/content/drive/My Drive/local_mt/fine_tuned_improved/data/vocab/amis-expanded.csv'
output_file_path = '/content/drive/My Drive/local_mt/fine_tuned_improved/data/vocab/amis-wiki-cleaned.csv'

clean_csv(input_file_path, output_file_path)
```
