```python
import csv
```

```python
# Saves text to files
class FileHandler:
  def __init__(self, properties: dict = {}):
    self.properties = properties

  def save(self, content: list, destination: str):
    with open(destination, 'a', encoding='utf-8') as out_file:
      for line in content:
        out_file.write(line + '\n')

  def read_csv(self, file: str):
    corpus = []
    with open(file) as csvfile:
      reader = csv.DictReader(csvfile)
      for row in reader:
        corpus.append(row)

    return corpus

  def write_csv(self, fields: list[str], rows: list[dict], location: str):
    try:
      with open(location, 'w') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=fields)
        writer.writeheader()
        writer.writerows(rows)
    except:
      print("Not good.")
```
